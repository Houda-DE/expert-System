```python
import numpy as np


matrix = np.array([[5,5,4,0,0,1],
                   [4,3,3,2,2,1],
                   [2,1,2,3,2,2],
                   [5,3,5,3,4,3],
                   [4,4,3,2,3,2],
                   [2,0,1,3,1,1],
                   [3,3,4,2,4,4],
                   [1,2,1,4,3,3],
                   [0,1,0,3,1,0],
                   [2,0,1,3,1,0],
                   [1,2,1,1,0,1],
                   [4,2,4,2,1,2],
                   [3,2,3,3,2,3],
                   [1,0,0,3,2,2],
                   [2,1,1,2,3,2]])

correlation_matrix = np.corrcoef(matrix, rowvar=False)

print("Correlation matrix:")
print(correlation_matrix)
```

```python
eigenvalues, _ = np.linalg.eig(correlation_matrix)

inertia = eigenvalues / np.sum(eigenvalues)

cumulative_inertia = np.cumsum(inertia)

print("Eigenvalues:")
print(eigenvalues)
print("\nPercentage of inertia:")
print(inertia)
print("\nCumulative percentage of inertia:")
print(cumulative_inertia)
```

```python
eigenvalues, eigenvectors = np.linalg.eig(correlation_matrix)

idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

principal_components = np.dot(matrix, eigenvectors)

contributions = (eigenvalues / (matrix.shape[0] - 1)) * 100

representational_qualities = np.square(principal_components) / np.sum(np.square(matrix), axis=0)

print("Principal components:")
print(principal_components)
print("\nContributions to principal components (%):")
print(contributions)
print("\nRepresentational qualities of individuals:")
print(representational_qualities)
```

```python
individuals_in_first_factorial = principal_components[:, 0]

print("Individuals in the first factorial:")
print(individuals_in_first_factorial)
```
